```python
import serial
import time

end_time = time.time() + 5

samples = []
ser = serial.Serial('COM5')
print(ser.name)

while time.time() < end_time:
    x = ser.read(5)
    samples.append(x)

print(len(samples))
print(samples)
ser.close()
```
